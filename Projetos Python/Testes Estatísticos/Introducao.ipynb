# Conhecendo os dados e explorando

## Base de dados tmdb 5000

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
sns.set_theme()
ax = sns.distplot(tmdb.vote_average)
ax.set(xlabel = 'Nota Média', ylabel = 'Densidade')
ax.set_title("Média de votos do tmdb 5000")
```

```python
sns.set_theme()
ax = sns.distplot(tmdb.vote_average, norm_hist= False, kde = False)
ax.set(xlabel = 'Nota Média', ylabel = 'Frequência')
ax.set_title("Média de votos do tmdb 5000")
```

```python
ax = sns.boxplot(tmdb.vote_average, orient = 'h')
ax.set_title("Distribuição de nota média do tmdb 5000")
ax.set_xlabel("Nota média")
```

```python
tmdb[tmdb.vote_average == 0]
```

```python
tmdb_mais_10_votos = tmdb.query("vote_count >= 10")
tmdb_mais_10_votos.describe()
```

Visto que não é relevante observar filmes com pouquissimos votos, iremos analisar agora filmes com mais de 10 votos

```python
sns.set_theme()
ax = sns.distplot(tmdb_mais_10_votos.vote_average, norm_hist= False)
ax.set(xlabel = 'Nota Média', ylabel = 'Frequência')
ax.set_title("Média de notas do tmdb 5000 para filmes com mais de 10 votos")
```

```python
ax = sns.boxplot(tmdb_mais_10_votos.vote_average, orient = 'h')
ax.set_title("Distribuição de nota média do tmdb 5000 com mais de 10 votos")
ax.set_xlabel("Nota média")
```

## Comparando também o movielens

```python
notas = pd.read_csv('ml-latest-small/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId")['rating'].mean()
nota_media_por_filme.head()
```

```python
sns.set_theme()
ax = sns.distplot(nota_media_por_filme)
ax.set(xlabel = 'Nota Média', ylabel = 'Frequência')
ax.set_title("Média de notas do movielens 100k")
```

```python
qt_de_votos_por_filme = notas.groupby("movieId")['rating'].count()
qt_de_votos_por_filme[qt_de_votos_por_filme > 10]
```

```python
filmes_com_pelo_menos_10_votos = qt_de_votos_por_filme[qt_de_votos_por_filme > 10].index.values
filmes_com_pelo_menos_10_votos
```

```python
nota_media_filmes_min_10_votos = nota_media_por_filme[filmes_com_pelo_menos_10_votos]
nota_media_filmes_min_10_votos
```

```python
sns.set_theme()
ax = sns.distplot(nota_media_filmes_min_10_votos)
ax.set(xlabel = 'Nota Média', ylabel = 'Frequência')
ax.set_title("Média de notas do movielens 100k com mínimo 10 votos")
```

```python
nota_media_filmes_min_10_votos
```

```python
ax = sns.boxplot(nota_media_filmes_min_10_votos.values, orient = 'h')
ax.set_title("Distribuição de nota média do movielens com mais de 10 votos")
ax.set_xlabel("Nota média")
```

```python
ax = sns.distplot(nota_media_filmes_min_10_votos, hist_kws= {'cumulative' : True},
                 kde_kws= {'cumulative' : True})
ax.set(xlabel = 'Nota Média', ylabel = '% acumulada de filmes com mínimo 10 votos')
ax.set_title("")
```
