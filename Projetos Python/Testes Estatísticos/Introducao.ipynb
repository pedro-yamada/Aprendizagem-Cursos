### Intervalo de confiança da média com t e ztest

```python
from statsmodels.stats.weightstats import zconfint
```

### ztest

```python
zconfint(nota_media_filmes_min_10_votos)
```

Exige normalidade da distribuição dos meus dados, o teste z desta forma gartante um intervalo de confiança em 3.42 e 3.46

### ttest

```python
from statsmodels.stats.weightstats import DescrStatsW

descr_todos_com_10_votos = DescrStatsW(nota_media_filmes_min_10_votos)
descr_todos_com_10_votos.tconfint_mean()
```

amostra pequena: ttest 

amostra grande: ztest

## O problema de amostras pequenas

```python
np.random.seed(75241)

temp = notas1.rating.sample(frac = 1)

medias = [temp[0:i].mean() for i in range(1, len(temp))]

sns.lineplot(medias)
```

# Comparando duas amostras

```python
np.random.seed(75241)
temp = notas1.rating.sample(frac = 1)

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.43205)

    return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(1, len(temp))])

sns.lineplot(x =valores[:,0],y = valores[:,1])
sns.lineplot(x =valores[:,0],y = valores[:,2])

plt.hlines(y = 0.05, colors= 'g', xmin = 2, xmax = len(temp), linestyles= '--')
plt.annotate('$p_{value}$', xy = (170, 0.2), color = 'g', fontsize = 14)
```

```python
# Este intervalo de confiança da a tiferença entre as duas médias

zconfint(notas1.rating, notas.rating)
```

```python
# p_value da diferença é <= 0.05, que indica que são diferentes 3.762112778881965e-09
ztest(notas1.rating, notas.rating)
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toy_story = DescrStatsW(notas1.rating)

comparacao = descr_toy_story.get_compare(descr_todas_as_notas)
comparacao.summary()
```

Podemos observar o $p_{value}$ desta forma como = 0.000, ou seja são diferentesm além disso esse método já fornece os intervalos de confiança 0.280	0.559

## Comparação não paramétrica com Wilcoxon Ranksums

```python
from scipy.stats import normaltest

stat, p = normaltest(notas1.rating)
p
```

Como $p <= 0.05$, **NÃO PODEMOS** definir a distribuição como normal

```python
from scipy.stats import ranksums

stat, p = ranksums(x = notas1.rating, y = notas593.rating)
p
```

Como $p <= 0.05$, podemos concluir que as amostras vêm de uma mesma distribuição

Desta forma, podemos concluir, como as distribuições iguais, e como a nota média do filme 593 é maior que a do filme 1, que o filme 593 é estatisticamente preferido do que ao filme 1
